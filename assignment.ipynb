```python
import pandas as pd

df = pd.read_csv("input.csv")

data = pd.DataFrame(columns=['Employee Code', 'Manager Employee Code', 'Last Compensation', 'Compensation',
                           'Last Pay Raise Date', 'Variable Pay', 'Tenure in Org', 'Performance Rating',
                           'Engagement Score', 'Effective Date', 'End Date'])

my_list = []
l=0
for index, row in df.iterrows():
    employee_data = {
        'Employee Code': row["Employee Code"],
        'Manager Employee Code': row["Manager Employee Code"],
        'Last Compensation': "NaN",
        'Compensation': row["Compensation"],
        'Last Pay Raise Date': "NaN",
        'Variable Pay': 0,
        'Tenure in Org': "NaN",
        'Performance Rating': "NaN",
        'Engagement Score': "NaN",
        'Effective Date': row["Date of Joining"],
        'End Date': "NaN"
    }
    data = data.append(employee_data, ignore_index=True)

    sorted_map = {}
    for i in range(1, 3):
        compensation_col = f"Compensation {i}"
        review_col = f"Review {i}"
        engagement_col = f"Engagement {i}"

        if not pd.isna(row[compensation_col + " date"]):
            sorted_map[row[compensation_col + " date"]] = row[compensation_col]
        if not pd.isna(row[review_col + " date"]):
            sorted_map[row[review_col + " date"]] = row[review_col]
        if not pd.isna(row[engagement_col + " date"]):
            sorted_map[row[engagement_col + " date"]] = row[engagement_col]

    sorted_mp = dict(sorted(sorted_map.items(), key=lambda x: pd.to_datetime(x[0], format="%d-%M-%Y")))

    prev_values = employee_data.copy()

    last_comp=rows['Compensation']
    check=0
    for date, value in sorted_mp.items():
        rows = prev_values.copy()

        if value <= 5:
            rows['Performance Rating'] = value
        elif 5 < value < 10:
            rows['Engagement Score'] = value
        else:
            last_comp=rows['Compensation']
            rows['Compensation'] = value
            check=check+1

        rows['Effective Date'] = date

        if(check!=0):
          rows['Last Pay Raise Date']=row[f"Compensation {check} date"]
          rows['Last Compensation']=last_comp

        data = data.append(rows, ignore_index=True)
        prev_values = rows
    l=l+len(sorted_mp)+1
    my_list.append(l)


data['Effective Date'] = pd.to_datetime(data['Effective Date'], format='%d-%m-%Y')
data['Effective Date'] = pd.to_datetime(data['Effective Date'])
data['End Date'] = pd.to_datetime(data['End Date'])
data['End Date'] = data['Effective Date'].shift(-1) - pd.Timedelta(days=1)
data['End Date'] = data['End Date'].dt.strftime('%Y-%m-%d')
ll=0
for i in my_list:
    if pd.isnull(df.loc[ll,"Date of Exit"]):
        data.loc[i-1, 'End Date'] = "2100-01-01"
    else:
        data.loc[i-1, 'End Date'] = pd.to_datetime(df.loc[ll, "Date of Exit"]).strftime("%Y-%m-%d")
    ll=ll+1

print(data)
```
